```python
# one shot prompt
!pip install transformers
from transformers import pipeline
translator= pipeline("translation_en_to_fr",model="Helsinki-NLP/opus-mt-en-fr")
prompt = "hi how are you"
print(translator(prompt))
```

```python
#chain of tought
from transformers import pipeline
cal= pipeline("text2text-generation",model="google/flan-t5-large")
prompt = """
ram has 99 apples , ram gave 10 apples to Krishna , so then how many apples are now at ram ? """
print(cal(prompt))
```

```python
# few  shot prompt
!pip install transformers
from transformers import pipeline
translator= pipeline("translation_en_to_es",model="google/flan-t5-base")
prompt = """ Translate English to Spanish:
English:Good morning
spanish:Buen dia
English : thank you
spanish :               """
print(translator(prompt))
```
